```python
import pandas as pd
import numpy as np
```

```python
df_student = pd.read_csv("student.csv")
df_student
```

#เปลี่ยนข้อมูลชื่อที่เป็น NaN ให้เป็นไม่ระบุ

```python
df_student['ชื่อ'].fillna('ไม่ระบุ', inplace=True)
```

#แปลงข้อมูล gpa ให้เป็น int แล้วกำหนดให้น้อยกว่า 0 กลายเป็น NaN แต่ถ้าเกินจะให้เป็น 4.0 และแปลงค่า NaN ให้เป็นค่าเฉลี่ยของเกรด

```python
df_student['GPA'] = pd.to_numeric(df_student['GPA'], errors='coerce')
df_student.loc[df_student['GPA'] < 0, 'GPA'] = np.nan
df_student.loc[df_student['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df_student['GPA'].mean()
df_student['GPA'].fillna(mean_gpa, inplace=True)
df_student['GPA'] = df_student['GPA'].round(2)
```

#ถ้าข้อมูลของแถว "คณะ" เขียนเหมือนฝั่งซ้ายจะให้แปลงเป็นฝั่งขวา

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'Science' : 'วิทยาศาสตร์',
    'Sci' : 'วิทยาศาสตร์',
    'คณะวิทยาศาสตร์' : 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'วิศวฯ' : 'วิศวกรรมศาตร์',
    'วิศวะ' : 'วิศวกรรมศาสตร์',
    'วิศวกรรม' : 'วิศวกรรมศาสตร์',
    'Engineering' : 'วิศวกรรมศาสตร์',
    'วิศวกรรมศาสตร์' : 'วิศวกรรมศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'ศิลป์' : 'ศิลปศาสตร์',
    'Liberal Arts' : 'ศิลปศาสตร์'
}
df_student['คณะ'] = df_student['คณะ'].map(faculty_mapping)
```

```python
mode_fault = df_student['คณะ'].mode()[0]
df_student['คณะ'].fillna(mode_fault, inplace=True)
```

#ทำให้เพศตรงตามมาตรฐาน

```python
Gender_mapping = {
    'ชาย' : 'ชาย',
    'Male': 'ชาย',
    'ช' : 'ชาย',
    'M' : 'ชาย',
    'หญิง' : 'หญิง',
    'Female' : 'หญิง',
    'ญ': 'หญิง',
    'F' : 'หญิง',
}
df_student['เพศ'] = df_student['เพศ'].map(Gender_mapping)
```

ให้เพศที่เป็น NaN เป็นเพศที่มีเยอะ

```python
mode_sex = df_student['เพศ'].mode()[0]
df_student['เพศ'].fillna(mode_sex, inplace=True)
```

แปลงอายุที่สูงเกิน ให้กลายเป็น NaN และเปลี่ยนข้อมูลที่เป็น NaN ให้กลายเป็นอายุที่เป็นฐานนิยม

```python
mode_age = df_student['อายุ'].mode()[0]
df_student.loc[df_student['อายุ'] > 22.0, 'อายุ'] = np.nan
df_student['อายุ'].fillna(mode_age, inplace=True)
```

แปลงค่าที่เป็น NaN ให้กลายเป็น ค่าฐานนิยมของ ส่วนสูง,น้ำหนัก และชั่วโมงเรียนต่อสัปดาห์

```python
mode_height = df_student['ส่วนสูง_cm'].mode()[0]
df_student['ส่วนสูง_cm'].fillna(mode_height, inplace=True)
mode_weight = df_student['น้ำหนัก_kg'].mode()[0]
df_student['น้ำหนัก_kg'].fillna(mode_weight, inplace=True)
mode_hours = df_student['ชั่วโมงเรียน_ต่อสัปดาห์'].mode()[0]
df_student['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mode_hours, inplace=True)
```
